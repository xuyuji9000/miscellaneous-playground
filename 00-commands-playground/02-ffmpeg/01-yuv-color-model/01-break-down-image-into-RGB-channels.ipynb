```python
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Prepare the image

img = np.asarray(Image.open('./The-Starry-Night.jpg'))
img.shape
```

```python
# Break the image down into Red, Green, Blue channels

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
# for ax in axs.flat:
#     ax.axis('off')
axs[0,0].imshow(img[:, :, 0])
axs[0,0].axis('off')
axs[0,0].set_title('Red Channel')

axs[0,1].imshow(img[:, :, 1])
axs[0,1].axis('off')
axs[0,1].set_title('Green Channel')

axs[1,0].imshow(img[:, :, 2])
axs[1,0].axis('off')
axs[1,0].set_title('Blue Channel')

axs[1,1].imshow(img[:, :, :])
axs[1,1].axis('off')
axs[1,1].set_title('Original Picture')
```
